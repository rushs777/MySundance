```python
import matplotlib.pyplot as plt
from IPython.display import display
import sys
import pandas as pd
```

```python
# Add the Results folders of the different components
# Forward Problem
FP_dir = '/home/sirush/PhDResearch/MMS_Transient_Channel/ForwardProblem/Log_Files/'
figures_dir = '/home/sirush/Dropbox/Research/PhD/LatexFiles/Dissertation/Figures/'
filename = 'Viento_Verification_Transient_Channel.png'
```

```python
# Begin by reading in the data from the forward problem
viento_verification_log = 'log_Viento_verification_short.txt'
viento_verification_data = pd.read_csv(FP_dir+viento_verification_log, sep=' ')

# Filter out the nx=128 case
viento_verification_data = viento_verification_data[viento_verification_data['nx'].ne(128)]
```

```python
plt.figure(1,figsize=(5,5))#.subplots_adjust(hspace=2)
for Re in viento_verification_data['Re'].unique():
    plt.plot(viento_verification_data['nx'].unique(),
             viento_verification_data[viento_verification_data['Re'].eq(Re)]['uErr'],
            label='Re={}'.format(Re))
# plt.plot(viento_verification_data['nx'].unique(),viento_verification_data[viento_verification_data['Re'].eq(1)]['uErr'])

plt.xlabel('Spatial & Temporal Resolution',weight='medium',size=16,labelpad=20)
plt.ylabel('$||u_{Ex}- u_{Viento}||$',weight='medium',size=20,labelpad=20)
# plt.title('Accuracy of Viento')
plt.xticks(viento_verification_data['nx'].unique(),size=16)
# plt.ylim(0, 1.1*viento_verification_data['uErr'].max())
# plt.ylim(0, 0.0002)
plt.yscale('log')
plt.yticks(size=16)
# plt.legend(loc='upper right')
plt.figlegend(loc='center',bbox_to_anchor=(1.3,0.5),fontsize=20)

# Either this or plt.gcf() for get current figure and then intearct with the Figure object
plt.savefig(figures_dir+filename,bbox_inches='tight', pad_inches=0.1)
# plt.savefig(filename, bbox_inches='tight')






plt.show() # Creates a new figure
```

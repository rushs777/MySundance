```python
import matplotlib.pyplot as plt
from IPython.display import display
import sys
import pandas as pd
```

```python
# Specify global paths
figures_dir = '/home/sirush/Dropbox/Research/PhD/LatexFiles/Dissertation/Figures/'
fileprefix = 'sensor_grid_Transient_Channel_numSensors'
```

```python
import numpy as np
# evenly sampled time at 200ms intervals

# t = np.arange(deltaX,1-deltaX,deltaX)

title_offset = 1.1
sensor_array = {1:'y',2:'r',3:'k',5:'c',11:'g'}
# sensor_array = [1]
for k in sensor_array.keys():
    deltaX = 1.0/(k+1)
    for i in range(1,k+1):
        x = deltaX*i
        for j in range(1,k+1):
            y = deltaX*j
    #         display('({},{})'.format(x,y))
            plt.scatter(x,y,marker='o',color=sensor_array[k],s=200)
    plt.axis([0,1,0,1])
    plt.xticks(size=18)
    plt.yticks(size=18)
    if k==1:
        plt.title('Refered to as ODECO with {} sensor'.format(k),weight='bold',size=20,y=title_offset)
    else:
        plt.title('Refered to as ODECO with {} sensors'.format(k),weight='bold',size=20,y=title_offset)
    plt.savefig(figures_dir+fileprefix+str(k)+'.png',bbox_inches='tight')
    plt.show()
    
# plt.savefig(figures_dir+absolute_filename, 
#             bbox_inches='tight', pad_inches=0.1)
```

```python
sensor_array = [1,2,3,5,11]
# sensor_array = [1]
for k in sensor_array:
    df = pd.DataFrame(columns=['x','y'])
    deltaX = 1.0/(k+1)
    for i in range(1,k+1):
        x = deltaX*i
        for j in range(1,k+1):
            y = deltaX*j
            df = df.append(pd.DataFrame(columns=['x','y'],data=[[x,y]]))
    df.to_csv('numSens={}.csv'.format(k),index=False)
```

```python
import matplotlib.pyplot as plt
from IPython.display import display
import sys
import pandas as pd
```

```python
# Add the Results folders of the different components
# Forward Problem
FP_dir = '/home/sirush/PhDResearch/MMS_Transient_Channel/ForwardProblem/Log_Files/'
```

```python
# Begin by reading in the data from the forward problem
forward_problem_log = 'log_Re{1,20,40,60,80,100}_nxnt{24,36,48,60,72}_short.txt'
forward_probem_data = pd.read_csv(FP_dir+forward_problem, sep=' ')
```
